### Featureの意味 メモ
- PE  
PE(Price to Earning Ratio)は 株価と利益の比率
PE := StockPrice / 1株当たり利益(EPS)  
高ければ割高株、低ければ割安株.  
traillingは直近1年の平均, forwardの場合は予想利益

- EPS  
EPS(Earnings Per Share)は 1株あたりの利益
EPS := 当期純利益 ÷ 発行済株式総数

### Feature一覧

```python
# categorical
cols_obj = jp.dtypes.to_frame("dtype").astype(str).query("dtype == 'object'").index.sort_values()
cols_obj
```

```python
# numeric
cols_num = jp.dtypes.to_frame("dtype").astype(str).query("dtype == 'float64'").index.sort_values()
cols_num
```

```python
# 全行Nullのカラム
cols_na = (jp.isna().sum(axis=0) == jp.shape[0]).to_frame("fullNa").query("fullNa==True").index
cols_na.sort_values()
```

```python
set(cols_na) - set(cols_num)
```

全てのcompanyでNAのカラムは全て numericなカラム

```python
# 全行NonNullのカラム
cols_full = (jp.isna().sum(axis=0) == 0).to_frame("NoNa").query("NoNa==True").index.sort_values()
cols_full
```

```python
# 非NULLのカラム
cols_num = pd.Index(set(cols_num) - set(cols_na)).sort_values()
cols_num
```
